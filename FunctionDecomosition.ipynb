```python
import numpy as np
import matplotlib.pyplot as plt

# Define the original function f(x)
def f(x):
    return x**2 - 4

# Define the positive part f_+(x)
def f_plus(x):
    return np.maximum(f(x), 0)

# Define the negative part f_-(x)
def f_minus(x):
    return np.maximum(-f(x), 0)

# Define the positive part f_>1(x)
def f_up1(x):
    return np.maximum(f(x), 1)

# Define the negative part f_<1(x)
def f_down1(x):
    return np.maximum(2-f(x), 1)

# Define the x range for plotting
x = np.linspace(-5, 5, 400)

# Calculate values for f(x), f_+(x), and f_-(x)
y = f(x)
y_plus = f_plus(x)
y_minus = f_minus(x)

# Calculate values for f(x), f_up(x), and f_down(x)
y = f(x)
y_up1 = f_up1(x)
y_down1 = f_down1(x)
rekon1=y_up1+(2-y_down1)-1
```

```python
# Create the plot
plt.figure(figsize=(12, 8))

# Plot f(x)
#plt.plot(x, y, label=r'$f(x) = x^2 - 4$', color='blue')

# Plot f_+(x)
#plt.plot(x, y_plus, label=r'$f_+(x) = \max(f(x), 0)$', color='green', linestyle='--')
plt.plot(x, y_plus, color='blue', linestyle='-', linewidth=3)

# Plot f_-(x)
#plt.plot(x, y_minus, label=r'$f_-(x) = \max(-f(x), 0)$', color='red', linestyle='--')
plt.plot(x, y_minus, color='red', linestyle='--', linewidth=3)

# Add titles and labels
#plt.title(r'Decomposition of Function $f(x)$ into $f_+(x)$ and $f_-(x)$')
#plt.xlabel(r'$x$')
#plt.ylabel(r'$y$')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid()
plt.legend()
plt.xlim(-5, 5)
plt.ylim(-5, 10)

# Show the plot
plt.show()
```
